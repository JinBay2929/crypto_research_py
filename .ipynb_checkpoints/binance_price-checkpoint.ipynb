```python
import requests

def get_current_prices():
    base_url = "https://api.binance.com"
    endpoint = "/api/v3/ticker/price"
    url = base_url + endpoint

    response = requests.get(url)
    data = response.json()

    prices = {item['symbol']: float(item['price']) for item in data}
    return prices

current_prices = get_current_prices()
print(current_prices)
```

```python
import requests

def get_uniswap_price(token_address):
    url = "https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2"
    query = """
    {
      token(id: "%s") {
        symbol
        derivedETH
      }
      bundle(id: "1") {
        ethPrice
      }
    }
    """ % token_address

    response = requests.post(url, json={"query": query})
    data = response.json()

    token_data = data["data"]["token"]
    eth_price = float(data["data"]["bundle"]["ethPrice"])
    derived_eth = float(token_data["derivedETH"])

    token_price = derived_eth * eth_price
    return token_data["symbol"], token_price

# Sample: Wrapped Ether (WETH) contract address on Ethereum mainnet
token_address = "0xc02aaa39b223fe8d0a0e5c4f27ead9083c756cc2"

symbol, price = get_uniswap_price(token_address)
print(f"{symbol}: ${price:.2f}")
```

```python
url = "https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2"

# GraphQL query to fetch the top 10 pairs by volume
query = """
{
  pairs(first: 10, orderBy: volumeUSD, orderDirection: desc) {
    id
    token0 {
      id
      symbol
    }
    token1 {
      id
      symbol
    }
    volumeUSD
  }
}
"""

response = requests.post(url, json={'query': query})

if response.status_code == 200:
    data = response.json()
    pairs = data['data']['pairs']

    print("Top 10 Uniswap V2 Pairs by Volume:")
    for i, pair in enumerate(pairs, start=1):
        token0_symbol = pair['token0']['symbol']
        token1_symbol = pair['token1']['symbol']
        volume_usd = float(pair['volumeUSD'])

        print(f"{i}. {token0_symbol}-{token1_symbol}: ${volume_usd:,.2f}")
else:
    print(f"Error: {response.status_code} - {response.text}")
```

```python
# Binance API endpoint for order book depth
url = "https://api.binance.com/api/v3/depth"

# Define the symbol for which you want to get the order book
symbol = "BTCUSDT"

# Define the parameters for the request
params = {
    "symbol": symbol,
    "limit": 5  # You can choose a different limit value (5, 10, 20, 50, 100, 500, 1000, 5000)
}

# Send the request
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON
    data = response.json()

    # Extract the bids (buy orders) and asks (sell orders)
    bids = data["bids"]
    asks = data["asks"]

    # Get the best bid and ask prices
    best_bid_price = float(bids[0][0])
    best_ask_price = float(asks[0][0])

    # Print the best bid and ask prices
    print(f"Best Bid Price for {symbol}: {best_bid_price:.2f}")
    print(f"Best Ask Price for {symbol}: {best_ask_price:.2f}")
else:
    print(f"Error: {response.status_code} - {response.text}")
```
